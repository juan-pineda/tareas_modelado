## El juego de la vida
Las reglas de este juego son las siguientes:

    1. Una persona nace si tiene alrededor cierta cantidad de personas que depende de la tasa de natalidad
    2. Una persona muere si tiene alrededor cierta cantidad de personas que depende de la tasa de mortalidad
    3. Una persona sobrevive si tiene alrededor cierta cantidad de personas que depende de la tasa de supervivencia
