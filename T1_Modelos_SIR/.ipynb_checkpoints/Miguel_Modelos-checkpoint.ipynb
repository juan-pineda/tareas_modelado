## Modelo SIRM

<p>Tomamos las ecuaciones del modelo SIR solo que esta vez agregaremos el parametro de las personas muertas, cabe aclarar que este modelo toma que una persona infectada que se recupera no se vulve a enfermar</p>

<div style="text-align: center">  $\frac {d}{dt}(S+I+R+M) = 0$ </div>
<div style="text-align: center">  $\frac {dS}{dt}= - \beta IS$ </div>
<div style="text-align: center">  $\frac {dR}{dt}=  \gamma I$ </div>
<div style="text-align: center">  $\frac {dI}{dt}=  \beta IS - \phi IS - \gamma I$  </div>
<div style="text-align: center">  $\frac {dM}{dt}=   \phi I$ </div>

<p>Beta hace referencia a la tasa de de infección</p>
<p>Gamma hace referencia a la tasa de recuperación</p>
<p>Phi hace referencia a la tasa de mortalidad</p>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Aca se calculan con beta,gamma,phi y el paso de tiempo los valores de Infectados,Suceptibles,Recuperados,Muertos
def new(beta,gamma,phi,I,S,R,M,dt):
    
#La razon de cambio de los recuperados depende de un Gamma y los infectados todo es respecto al paso del tiempo
    dR = (gamma*I)*dt
    R_new = R + dR
#La razon de cambio de los suceptibles depende de un Beta que va a ser negativo ya que tienden a volverse infectados
    dS = (-beta*I*S)*dt
    S_new = S + dS
#La razon de cambio de las muertes solo depende de un Phi y de los infectados
    dM = (phi*I)*dt
    M_new = M + dM
    I_new = 1 - S_new - R_new - M_new
    return I_new, S_new, R_new, M_new
```

```python
#Se definen las funciones a graficar
def dibujar(times,I,S,R,M):
    plt.figure()
    plt.plot(times,I,"r", label = "Infectados")
    plt.plot(times,S,"b", label = "Suceptibles")
    plt.plot(times,R,"g", label = "Recuperados")
    plt.plot(times,M,"k", label = "Fallecidos")
    plt.legend()
    
```

```python
def hagatodo (beta,gamma,phi,tmax,dt):
    I = [0.01]
    S = [0.99]
    R = [0]
    M = [0]
    
    times = np.arange(0,tmax+dt,dt)
    
    for i in range(times.size-1):
        I_new, S_new, R_new, M_new = new(beta,gamma,phi,I[i],S[i],R[i],M[i],dt)
        I.append(I_new)
        S.append(S_new)
        R.append(R_new)
        M.append(M_new)
    dibujar(times,I,S,R,M)
    
    
```

```python
#Podemos observar que a medida que aumentamos el beta y se disminuye el gama el pico de infectados llega mas alto y mas rapido
hagatodo(beta=0.4,gamma=0.2,phi=0.02,tmax=100,dt=0.1)
```

```python
#se puede observar que si se controla el Beta podemos disminuir y se podria "aplanar la curva" para que no supere los sistemas de salud
hagatodo(beta=0.6,gamma=0.2,phi=0.02,tmax=100,dt=0.1)
```

```python
#la tasa de fallecidos depende de los infectados se puede observar  que el pico de infectados no estan alto 
#pero es debido a que se suman a la curva de muertes tambien influye que aca se aumenta la tasa de mortalidad del virus
#tambien observamos como la curva de suceptibles se aporxima a la curva de fallecidos
hagatodo(beta=0.8,gamma=0.3,phi=0.1,tmax=100,dt=0.1)
```
