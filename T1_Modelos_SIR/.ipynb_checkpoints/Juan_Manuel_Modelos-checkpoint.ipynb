En este código se modela el comportamiento pandémico de una enfermedad. 
El modelo empleado es el SIRD, en donde se considera inmunidad después de la enfermedad
y la muerte. Se asume una población dividida en los siguientes grupos:

S := compartimento de personas susceptibles. <br />
I := compartimento de personas infectadas (infecciosas). <br />
R := compartimento de personas recuperadas. <br />
D := compartimento de personas muertas. 

Los parámetros que caracterizan el comportamiento del modelo son:

$\beta$  := taza de infección.<br />
$\gamma$ := taza de recuperación espontánea.<br />
$\mu$    := taza de muerte.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def new(beta, gamma, mu, S, I, R, D, dt):
    
    """
    Esta función calcula el valor de las funciones en cada paso de tiempo dt.
    """
    
    dR = (gamma*I)*dt
    R_new = R + dR 

    dS = (-beta*I*S)*dt
    S_new = S + dS

    #El modelo del compartimento de fallecidos se empleó teniendo en cuenta que estos solo dependen del
    #número de personas infectadas, de modo que tiene un comportamiento análogo al número de recuperados. 
    
    dD = (mu*I)*dt
    D_new = D + dD
    
    I_new = 1 - S_new - R_new - D_new 
    
    return S_new, I_new, R_new, D_new
```

```python
def plotme(times, S, I, R, D,tmax,legend):
    """
    Esta función grafica las distintas curvas para los diferentes compartimentos.
    """
    
    plt.rc('text', usetex=True)
    plt.rc('font',family='serif')
    plt.style.use('seaborn-whitegrid')   
    plt.figure
    if (legend == True):
        plt.plot(times, S, "-b", label = r"Susceptibles")
        plt.plot(times, I, "--r", label = r"Infectados")
        plt.plot(times, R, ":g", label = r"Recuperados")
        plt.plot(times, D, "-.k", label = r"Muertos")
    else:
        plt.plot(times, S, "-b")
        plt.plot(times, I, "--r")
        plt.plot(times, R, ":g")
        plt.plot(times, D, "-.k")
    plt.xlabel(r"Tiempo", fontsize = 16)
    plt.ylabel(r"Población", fontsize = 16)
    plt.legend()
    plt.xlim(0, tmax)
    plt.ylim(0, 1)
```

```python
def main(beta, gamma, mu, tmax, dt,legend):
    """
    Esta función ejecuta todas las funciones. Es aquí en donde se realizan todas las iteraciones.
    """    
    S = [0.95]
    I = [0.05]
    R = [0]
    D = [0]
    
    times = np.arange(0,tmax+dt,dt)
    
    for i in range(times.size-1):
        
        S_new, I_new, R_new, D_new = new(beta, gamma, mu, S[i], I[i], R[i], D[i],dt)

        S.append(S_new)
        I.append(I_new)
        R.append(R_new)
        D.append(D_new)
        
    plotme(times, S, I, R, D,tmax,legend)
```

```python
#Para recuperar el modelo SIR original: mu = 0
#Para el Covid-19: beta = 2.5*gamma

main(beta=0.3125, gamma=0.125, mu = 1/30, tmax=100, dt=0.1, legend = True)
```

Para distintos valores de $dt$:

```python
main(beta=0.3125, gamma=0.125, mu = 1/30, tmax=100, dt=0.1, legend = True)
main(beta=0.3125, gamma=0.125, mu = 1/30, tmax=100, dt=1.0, legend = False)
```

Es claro que en el intervalo $[0.1, 1.0]$ las gráficas se distancian entre sí. Ahora observemos el intervalo $[0.01, 0.1]$:

```python
main(beta=0.3125, gamma=0.125, mu = 1/30, tmax=100, dt=0.01, legend = True)
main(beta=0.3125, gamma=0.125, mu = 1/30, tmax=100, dt=0.1, legend = False)
```

En este, es prácticamente imposible discernir entre ambas gráficas. Es decir, el método converge, por lo cual el paso de tiempo $dt = 0.1$ es suficiente. Observemos ahora el impacto de $\beta$ y $\gamma$ en la infección sin considerar la muerte $(\mu = 0)$:

```python
main(beta=0.3125, gamma=0.125, mu = 0, tmax=100, dt=0.1, legend = True)
main(beta=0.625, gamma=0.25, mu = 0, tmax=100, dt=0.1, legend = False)
```

Al considerar una mayor taza de contagiados con base en la relación $\beta = 2.5\mu$, es claro que las cotas (asíntotas para el compartimento de recuperados y de susceptibles) son las mismas, a pesar de que hay un desplazamiento hacia la izquierda con respecto a la grafica original (curvas a la derecha). Esto indica que al aumentar el número de contagiados alcanzará el pico de infección más rápido, pero así mismo también desaparecerá más pronto. Bajo este escenario lo recomendable sería un contagio voluntario. Ahora, al considerar la taza de muertos del $10 \%$:

```python
main(beta=0.3125, gamma=0.125, mu = 1/10, tmax=100, dt=0.1, legend = True)
```

```python
main(beta=0.625, gamma=0.25, mu = 1/10, tmax=100, dt=0.1, legend = True)
```

```python
main(beta=0.3125, gamma=0.125, mu = 1/10, tmax=100, dt=0.1, legend = True)
main(beta=0.625, gamma=0.25, mu = 1/10, tmax=100, dt=0.1, legend = False)
```

Con base en esta gráfica es posible observar que si bien con una taza de infección mayor se llega al pico de infección más rápido y es más alto en relación a una taza más baja, el número de personas fallecidas es menor y alcanza su asíntota más rápido. En conclusión, en este modelo una una enfermedad altamente contagiosa se ve reflejada en un número de muertes menor. Ahora, con una taza de mortalidad del $3\%$:

```python
main(beta=0.3125, gamma=0.125, mu = 3/100, tmax=100, dt=0.1, legend = True)
main(beta=0.625, gamma=0.25, mu = 3/100, tmax=100, dt=0.1, legend = False)
```
