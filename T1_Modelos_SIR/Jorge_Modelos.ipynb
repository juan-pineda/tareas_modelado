```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')
```

## Modelo SIR (con nacimiento, muerte natural y muerte debido a enfermedad)


Se considera una población fija con sólo tres estados: susceptibles $(S)$, infectados $(I)$ y resistentes $(R)$. Este modelo resulta adecuado cuando los agentes infecciosos son virus. Con $S(t)$, $I(t)$ y $R(t)$ las variables que representan el número de individuos de cada clase, se asume que:

1. El número de infectados aumenta a una tasa (tasa de infección) $\beta > 0$, proporcional al número de infectados y al número de susceptibles: $\beta I S$. Además, se supone que el número de susceptibles disminuye acorde con esta tasa.

2. La tasa de recuperación, $\gamma > 0$, es proporcional al número de infectados: $\gamma I$. Además, el número de resistentes aumenta acorde con esta tasa.

3. El tiempo de incubación es despreciable: un susceptible, al infectarse, inmediatamente se vuelve infeccioso.

Considerando una población fija, $N$, se tiene:

$$S(t) + I(t) + R(t) = 1,$$

con $S(t) = S/N$, $I(t) = I/N$, y $R(t) = R/N$, las variables normalizadas. Así, el sistema de ecuaciones diferenciales que describe el modelo está dado por:


$$\displaystyle{\dfrac{dS}{dt} = - \beta I S},$$

$$\displaystyle{\dfrac{dI}{dt} = \beta I S - \gamma I},$$

$$\displaystyle{\dfrac{dR}{dt} = \gamma I}.$$

Se además se considera la muerte por enfermadad que sufren los individuos infectados (con el fin de mantener la población constante, se considerará una cantidad igual de nacidos y fallecidos), el modelo transforma en:

$$\displaystyle{\dfrac{dS}{dt} = - \beta I S + \alpha(1 - S) + \gamma \, \theta I},$$

$$\displaystyle{\dfrac{dI}{dt} = \beta I S - \alpha I - \theta \, \gamma I - (1 - \theta) \, \gamma I = \beta I S - \alpha I - \gamma I},$$

$$\displaystyle{\dfrac{dR}{dt} = (1 - \theta) \, \gamma I - \alpha R},$$

donde $\beta$ es la tasa de infección, $\gamma$ es la tasa de recuperación, $\alpha$ es la tasa de natalidad/mortalidad por causas naturales, y el parámetro $\theta$ representa el número de individuos que mueren a causa de la enfermedad. 


## Modelo SIRS


$$\displaystyle{\dfrac {dS}{dt}= - \beta I S + \mu R}$$

$$\displaystyle{\dfrac {dI}{dt} = \beta I S - \gamma I}$$

$$\displaystyle{\dfrac {dR}{dt} = \gamma I - \mu R}$$

## Solución numérica


$$x_{n + 1} = x_n + \Delta t \, f'(x_n)$$

```python
# Modelo SIR modificado

def SIR_M(S, I, R, dt, alpha, beta, gamma, theta):
    
    S_next = S + (dt * (- beta * I * S + alpha *(1 - S) + gamma * theta * I))
    I_next = I + (dt * (beta * I * S - gamma * I - alpha * I))
    R_next = R + (dt * ((1 - theta) * gamma * I - alpha * R))
    
    return S_next, I_next, R_next
```

```python
# Modelo SIRS

def SIRS(S, I, R, dt, beta, gamma, mu):
    
    S_next = S + (dt * (- beta * I * S + mu * R))
    I_next = I + (dt * (beta * I * S - gamma * I))
    R_next = R + (dt * (gamma * I - mu * R))
    
    return S_next, I_next, R_next
```

```python
# Función que calcula las soluciones para diferentes dt, y las compara

def steps(time, alpha, beta, gamma, theta):
    
    stp = [1.5, 1, 0.5, 0.1, 0.05, 0.01]
    
    j = 0
    
    SS = np.zeros((len(np.arange(0, time, np.amin(stp))),len(stp)))
    II = np.zeros((len(np.arange(0, time, np.amin(stp))),len(stp)))
    RR = np.zeros((len(np.arange(0, time, np.amin(stp))),len(stp)))
    
    fig, A = plt.subplots(1, 3, figsize = (16, 4))
    
    for dt in stp:
        
        points = np.arange(0, time, dt)
        
        k = 0
        
        S = [0.95]
        I = [0.05]
        R = [0.1]
        
        for i in range(len(points) - 1):
    
            S_next, I_next, R_next = SIR_M(S[i], I[i], R[i], dt, alpha, beta, gamma, theta)
    
            S.append(S_next)
            I.append(I_next)
            R.append(R_next)  
            
            SS[k, j] = S_next 
            II[k, j] = I_next
            RR[k, j] = R_next
                        
            k = k + 1
        
        aux = np.arange(0, time - dt, dt)
        
        A[0].plot(aux, SS[0:len(aux),j], label = r'$dt = {:.2f}$' .format(dt))
        A[1].plot(aux, II[0:len(aux),j], label = r'$dt = {:.2f}$' .format(dt))
        A[2].plot(aux, RR[0:len(aux),j], label = r'$dt = {:.2f}$' .format(dt))
        
        j = j + 1
    
    plt.sca(A[0])
    plt.title (r'S Solution', fontsize = 14, pad = 12)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.ylabel(r'Subpopulations $(t)$', fontsize = 16, labelpad = 12)
    
    plt.sca(A[1])
    plt.xlabel(r'$t$', fontsize = 16, labelpad = 12)
    plt.title (r'I Solution', fontsize = 14, pad = 12)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.xlabel(r'$t$', fontsize = 14, labelpad = 12)
    
    plt.sca(A[2])
    plt.title (r'R Solution', fontsize = 14, pad = 12)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    
    plt.legend(frameon = False, fontsize = 14)
    fig.tight_layout()
    
    #plt.savefig('test3.eps')
    plt.show()
```

```python
# Función para graficar

def plot(S, I, R, points, alpha, beta, gamma, theta, mu):

    plt.figure(figsize=(8, 5))

    plt.plot(points, S, 'orange', label = 'S', lw = 2, ls = ':')
    plt.plot(points, I, 'blue', label = 'I', lw = 2, ls = '--')
    plt.plot(points, R, 'green', label = 'R', lw = 2, ls = '-')
    
    #plt.suptitle(r'Modelo SIR con nacimientos, muerte natural, y por enfermedad', fontsize = 16)
    plt.title(r'$\beta = {:.3f}, \,\, \gamma = {:.3f}, \,\, \alpha = {:.3f}, \,\, \theta = {:.3f}, \,\, \mu = {:.3f}.$' .format(beta, gamma, alpha, theta, mu), fontsize = 16, pad = 12)
    plt.xlabel(r'$t$', fontsize = 16, labelpad = 12)
    plt.ylabel(r'Subpopulations $(t)$', fontsize = 16, labelpad = 12)
    plt.xticks(fontsize = 16)
    plt.yticks(fontsize = 16)
    plt.legend(frameon = False, fontsize = 16)
    
#    plt.legend(loc = 'center right', frameon = False, fontsize = 16)

    plt.savefig('test1.png')

    plt.show()
```

```python
# Función para llamar todo

def call_all(model, time, dt, alpha, beta, gamma, theta, mu):
    
    S = [0.95]
    I = [0.05]
    R = [0.1]
    
    points = np.arange(0, time, dt)
        
    for i in range(len(points) - 1):
    
        if model == 'SIR_M':
    
            S_next, I_next, R_next = SIR_M(S[i], I[i], R[i], dt, alpha, beta, gamma, theta)
        
        elif model == 'SIRS':
            
            S_next, I_next, R_next = SIRS(S[i], I[i], R[i], dt, beta, gamma, mu)
    
        S.append(S_next)
        I.append(I_next)
        R.append(R_next)
    
    plot(S, I, R, points, alpha, beta, gamma, theta, mu)
```

# Estudio del paso ($dt$)

```python
steps(time = 50, alpha = 1/8, beta = 6/8, gamma = 1/8, theta = 1/8)
```
